```python
class NeuralNetwork(object):
    def __init__ (self, initDimen):
        self.length = len(initDimen)  
        
        self.inputLayer = initDimen[0]
        
        self.hiddenLayer = []

        for i in range(1, self.length-1): 
            self.hiddenLayer.append(initDimen[i])

            
        self.outputLayer = initDimen[self.length-1]

        self.w = []
    
        
        self.w.append(np.random.randn(self.inputLayer, self.hiddenLayer[0]))

        
        for i in range(1, self.length -2):
            self.w.append(np.random.randn(self.hiddenLayer[i-1], self.hiddenLayer[i]))
 
        
        self.w.append(np.random.randn(self.hiddenLayer[len(self.hiddenLayer)-1], self.outputLayer))

        
    def forward(self, x):
        self.hiddenVal = []
        self.nodeVal = []
        
        self.hiddenVal.append(np.dot(x,self.w[0]))
        self.nodeVal.append(self.sigmoid(self.hiddenVal[0]))
 
        for i in range(1, self.length - 2):
            
            self.hiddenVal.append(np.dot(self.nodeVal[i-1], self.w[i]))
            self.nodeVal.append(self.sigmoid(self.hiddenVal[i-1]))

        
        self.hiddenVal.append(np.dot(self.nodeVal[self.length - 3], self.w[self.length - 2]))
        self.nodeVal.append(self.sigmoid(self.hiddenVal[self.length - 2]))
      
        self.yHat = self.nodeVal[self.length - 2]   
           
        return self.yHat
    
    def sigmoid(self, x):
        return 1/(1+np.exp(-x))
    
    def sigmoidPrime(self, x):
        return np.exp(-x)/((1+np.exp(-x))**2)
    
    def residual(self, x, y):
        self.yHat = self.forward(x)
        r = 0.5 * sum((y*100-self.yHat*100)**2)
        return r
    
    def residualPrime(self, x, y):
        self.yHat = self.forward(x)
        
        resi = -(y - self.yHat)
        
        temp = [0] * (self.length - 1)
        self.delta = [0] * (self.length - 1)
        
        temp[self.length - 2] = np.multiply(resi, self.sigmoidPrime(self.hiddenVal[self.length - 2]))
        self.delta[self.length - 2] = np.dot(self.nodeVal[self.length -3].T, temp[self.length - 2])
        # print "temp :"
        #print temp[self.length - 2]
        #print "delta:"
        #print self.delta[self.length - 2]

        for i in reversed(range(0,self.length -2)):

            temp[i] = np.dot(temp[i+1], self.w[i+1].T) * self.sigmoidPrime(self.hiddenVal[i])
            self.delta[i] = np.dot(self.nodeVal[i-1].T, temp[i])
            #print "temp :"
            #print temp[i]
            #print "delta:"
            #print self.delta[i]
           
        
        self.delta[0] = np.dot(x.T, temp[0])
       
        return self.delta
    
    def upgrade(self, scale):
        for i in range(0,len(self.w)):
            self.w[i] = self.w[i] - scale* self.delta[i]
                
```

```python
class EpicTrainer(object):
    def __init__(self, NN, x, y, x1, y1):
        self.NN = NN
        
        self.x = x
        self.y = y
        
        self.oracleX = x1
        self.oracleY = y1
        
        self.curResi = self.findResi()+10
        self.newResi = 1
        self.scale = 1
    
    def setScale(self, scale):
        self.scale = scale
    
    def findResi(self):
        return NN.residual(self.oracleX, self.oracleY)
    
    def storeCurVal(self):
        self.w = NN.w
    
    def getChange(self):
        self.delta = NN.residualPrime(self.x,self.y)
     
    def activateChange(self):
        NN.upgrade(self.scale)
        
    def revertChange(self):
        NN.w = self.w
        
    def train(self):
        self.storeCurVal()
        
        self.getChange()
        
        self.activateChange()
        
        self.newResi = self.findResi()
        print self.newResi
        
        if self.newResi > self.curResi:
            self.revertChange()
            return False
        else:
            self.curResi = self.newResi
        return True
        
        
```
