```python
class neuron(object):
    
    def __init__(self, DNA):
        self.DNA = DNA
        
    def fitness(self, heuristic):
        self.fitVal = 0
        for index in range(len(self.DNA)):
            if heuristic[index] is self.DNA[index]:
                self.fitVal +=1
        return self.fitVal
```

```python
class evolution(object):
    
#suggested population size bigger than 10 for 5 letter word
    
    def __init__(self, population, DNAlength):
        self.population = population
        self.popHalf = int(population/5)
        self.DNAlength = DNAlength
        self.DNAhalf = int(DNAlength/2)
        self.generateInit()
        
    def generateInit(self):
        self.cells = []
        for i in range(self.population):
            DNA = ""
            # the 5 in range needs to be change to something else that is more genearlized
            for i2 in range(self.DNAlength):
                DNA = DNA + random.choice(string.ascii_lowercase)
            #print DNA
            cell = neuron(DNA)
            self.cells.append(cell)
    
    def populationFitness(self, key):
        totalFitness = 0
        for cell in self.cells:
            totalFitness += cell.fitness(key)
        self.avgFitness = totalFitness *1.0/ len(self.cells)
            
    def weedPopulation(self):
        self.oldGeneration = []
        
        for cell in self.cells:
            if cell.fitVal > self.avgFitness:
                self.oldGeneration.append(cell)
                
        
        self.oldGeneration = sorted(self.oldGeneration, key = lambda cell: cell.fitVal, reverse= True)
        while len(self.oldGeneration) > self.popHalf:
            #print len(self.oldGeneration)
            del self.oldGeneration[-1]
            del self.oldGeneration[randint(2,len(self.oldGeneration)-1)]
        
        sumed = 0
        for cell in self.oldGeneration:
            sumed += cell.fitVal
        self.oldGenAVG = sumed*1.0/ len(self.oldGeneration)
        #self.oldGenAVG = reduce(lambda x,y: x.fitVal+y.fitVal, self.oldGeneration)/len(self.oldGeneration)
        #maybe add something here that increase size of oldGeneration by keeping exactly 50% or something
    
    #mutationLevel is a number between 0 and 1
    def breed(self, someMutation, mutationLevel):
        # types of things that happen when you breed
            #mutations
                #swap
                #rotate
                
            #crossovers / blending of genetics
        self.newGeneration = []
        self.newGeneration.extend(self.oldGeneration)
        for cell in self.oldGeneration:
            
            if random.random() > someMutation:
                DNA = list(cell.DNA)
                DNA[randint(0,len(DNA)-1)] = random.choice(string.ascii_lowercase)
                DNA = "".join(DNA)
                cell = neuron(DNA)
                self.newGeneration.append(cell)
                
            elif random.random() > mutationLevel:
                DNA = list(cell.DNA)
                for i in range(self.DNAhalf):
                    DNA[randint(0,len(DNA)-1)] = random.choice(string.ascii_lowercase)
                DNA = "".join(DNA)
                cell = neuron(DNA)
                self.newGeneration.append(cell)
        
        
        if len(self.newGeneration) < self.population:
            while len(self.newGeneration) < self.population:
                if random.random()>0.3:
                    #using just father and mother
                    father = self.newGeneration[randint(0,len(self.oldGeneration)-1)]
                    mother = self.oldGeneration[randint(0,len(self.oldGeneration)-1)]

                    if father is mother:

                        continue
                    child = []
                    for index in range(self.DNAlength):
                        if random.random()>0.5:
                            child.append(father.DNA[index])
                        else:
                            child.append(mother.DNA[index])
                    child = ''.join(child)
                    child = neuron(child)
                    self.newGeneration.append(child)

                    #should try and orgy instead where we mix up everyone's dna
                else:
                    self.newGeneration.append(self.cells[randint(0,len(self.cells)-1)])
        
        self.cells = self.newGeneration
```
