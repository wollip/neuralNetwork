```python
from random import randint, random
from operator import add
from functools import reduce
```

```python
def individual(length, min, max):
    #return ''.join(chr(i) for i in )
    return [ randint(min,max) for x in range(length) ]

def population(count, length, min, max):
    return [ individual(length, min, max) for x in range(count) ]

def grade(pop, target):
    summed = reduce(add, (fitness(x, target) for x in pop))
    return summed / (len(pop) * 1.0)
```

```python
def fitness(individual, target):
    #return fitness1(individual, target)
    return fitness2(individual, target)

def fitness1(individual, target):
    count = 0
    for i in range(len(individual)):
        if individual[i] == ord(target[i]):
            count += 1
    return count

def fitness2(individual, target):
    count = 0
    for i in range(len(individual)):
        count += abs(individual[i] - ord(target[i]))
    return count
```

```python
def evolve(pop, target, retain=0.2, random_select=0.05, mutate=0.01):
    graded = [ (fitness(x, target), x) for x in pop]
    graded = [ x[1] for x in sorted(graded)]
    #graded.reverse()
    retain_length = int(len(graded)*retain)
    parents = graded[:retain_length]
    # randomly add other individuals to
    # promote genetic diversity
    for individual in graded[retain_length:]:
        if random_select > random():
            parents.append(individual)
    # mutate some individuals
    for individual in parents:
        if mutate > random():
            pos_to_mutate = randint(0, len(individual)-1)
            # this mutation is not ideal, because it
            # restricts the range of possible values,
            # but the function is unaware of the min/max
            # values used to create the individuals,
            individual[pos_to_mutate] = randint(ascii_a, ascii_z)
    # crossover parents to create children
    parents_length = len(parents)
    desired_length = len(pop) - parents_length
    children = []
    while len(children) < desired_length:
        male = randint(0, parents_length-1)
        female = randint(0, parents_length-1)
        if male != female:
            male = parents[male]
            female = parents[female]
            half = int(len(male) / 2)
            child = male[:half] + female[half:]
            children.append(child)
    parents.extend(children)
    return parents
```

```python
def printPop(pop):
    output = [ ''.join(chr(i) for i in individual) for individual in pop ]
    print(output)
```

```python
#target = 'hello'
#target = 'medium'
#target = 'algorithm'
#target = 'hippopotomonstrosesquippedaliophobia'
target = 'mynameisjason'
ascii_a = 97
ascii_z = 122
pop = population(1000, len(target), ascii_a, ascii_z)
for i in range(0, 200):
    pop = evolve(pop, target, random_select=0.2, mutate=0.1)
    print('\nGeneration ', i+1)
    printPop(pop)
```
